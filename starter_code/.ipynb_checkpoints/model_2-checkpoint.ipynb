```python
import pandas as pd
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.
selected_features = df[['koi_period', 'koi_time0bk', 'koi_slogg', 'koi_srad', 'ra', 'dec', 'koi_kepmag']]
X = selected_features
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
y = df['koi_disposition']
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.utils import to_categorical
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)
X_train.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale your data
X_scalar = MinMaxScaler().fit(X_train)
X_train_scaled = X_scalar.transform(X_train)
X_test_scaled = X_scalar.transform(X_test)
```

# Train the Model



```python
from sklearn.preprocessing import LabelEncoder
# step 1: label-encode dataset
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train)
label_encoder.fit(y_test)
encoded_y_test = label_encoder.transform(y_test)
```

```python
for label, original_class in zip(encoded_y_train, y_train):
    print('Original Class: ' + str(original_class))
    print('Encoded Label: ' + str(label))
    print('-' * 12)
```

```python
y_train_categorical = to_categorical(encoded_y_train)
y_test_categorical = to_categorical(encoded_y_test)
```

```python
y_train_categorical
```

```python
# create a normal neural network with 2 input, 6 hidden notes, and 2 output
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense


model2 = Sequential()
model2.add(Dense(units=6, activation='relu', input_dim=7))
model2.add(Dense(units=3, activation='softmax'))
```

```python
model2.summary()
```

```python
# compile the model
model2.compile(optimizer='adam', 
              loss = 'categorical_crossentropy',
              metrics =['accuracy'])
```

```python
print(X_train_scaled.shape,
    y_train_categorical.shape)
```

```python
# fit the model to the training data
model2.fit(
    X_train_scaled,
    y_train_categorical,
    epochs = 100,
    shuffle = True,
    verbose = 2)
```

```python
model_loss, model_accuracy = model2.evaluate(
    X_test_scaled, y_test_categorical, verbose =2)
print(f"Normal Neral Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

# Save the Model
